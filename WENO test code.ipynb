```python
import pyweno

# Plot inline and make figures pretty
%matplotlib inline
rcParams['figure.figsize'] = [8, 6]
rcParams['font.size'] = 20.
rcParams['font.weight'] = 'bold'
rcParams['font.family'] = 'serif'
rcParams['font.sans-serif'] = 'serif'
rcParams['text.usetex'] = True
rcParams['lines.linewidth'] = 1.5
```

```python
# Grid
t0 = 0; t = 0.1
x0 = 0.; x1 = 1.
N1 = 512; NG = 3
dx = (x1 - x0)/N1
x = zeros([N1+2*NG])
x[NG:-NG] = linspace(x0, x1, N1)

# Wavespeed 
c = 1.
courant = 0.1
dt = courant*dx/abs(c)
numTimeSteps = (int)((t - t0)/dt)

# Initial conditions
amplitude = 0.1
sigma = 0.1
xc = (x0 + x1)/2.
u = zeros([x.size, numTimeSteps])
u[NG:-NG, 0] = amplitude*exp(-(x[NG:-NG]-xc)**2./sigma**2.)
```

```python
def riemannSolver(fluxL, fluxR, uL, uR):
    return 0.5*(fluxL + fluxR) - 0.5*dx/dt*(uR - uL)
    
def setBoundaries(u):
    # Periodic boundary conditions
    u[:NG] = u[-2*NG:-NG]
    u[-NG:] = u[NG:2*NG]
    
def timeStep(u, n):
    
    setBoundaries(u[:, n])
    
    # Get left and right states
    uL = pyweno.weno.reconstruct(u[:, n], 5, 'left')
    uR = pyweno.weno.reconstruct(u[:, n], 5, 'right')
    
    flux = riemannSolver(c*uR[1:], c*uL[:-1], uR[1:], uL[:-1])
    
    u[:-2, n+1] = u[:-2, n] - dt/dx*(flux[1:] - flux[:-1]) 
    
```

```python
for n in xrange(numTimeSteps):
    print "Time step = ", n
    timeStep(u, n)
```

```python
plot(x[NG:-NG], u[NG:-NG, 0])
plot(x[NG:-NG], u[NG:-NG, 100])
plot(x[NG:-NG], u[NG:-NG, 200])
plot(x[NG:-NG], u[NG:-NG, 300])
```
