```python
import glob
import h5py

#filepath = '/Users/francoisfoucart/Work/grim/Runs/torus/TwoLoops/NoHO-Vm0/'
filepath = '/tmp/'


data_files = np.sort(glob.glob(filepath+'/data*.h5'))[:]

N1 = 128
N2 = 128
dof = 9
NG = 3
A_SPIN = 0.9375
H_SLOPE = 0.3
R_IN = 0.98*(1. + sqrt(1. - A_SPIN**2.))
R_OUT = 40.
X1_START = log(R_IN)
X2_START = 1e-20
DX1 = (log((R_OUT)/(R_IN))/N1)
DX2 = ((1. - 2.*X2_START)/N2)
i = arange(-NG, N1+NG, 1)
j = arange(-NG, N2+NG, 1)
X1 = (X1_START + (i + 0.5)*DX1)
X2 = (X2_START + (j + 0.5)*DX2)
R = exp(X1)
THETA = pi*(X2) + ((1 - H_SLOPE)/2.)*sin(2.*pi*(X2))

# Load the primitive variables
data = zeros([N2, N1, dof, data_files.size]) 
for i, dumpfile in enumerate(data_files):
    print "i = ", i
    datafile = h5py.File(dumpfile, "r")
    data[:, :, :, i] = datafile[datafile.items()[0][0]][:]
    datafile.close()
    
# Load reference data
#dataref = zeros([N2, N1, 8])
#datafile = h5py.File("/Users/francoisfoucart/Work/grim/Runs/torus/Viscosity/T1300-V0/data001300.h5", "r")
#dataref[:, :, :] = datafile[datafile.items()[0][0]][:]
#datafile.close()
    
# Load the geometry
gcov = zeros([N2, N1, 4, 4])
gcon = zeros([N2, N1, 4, 4])
gammaDownDownDown = zeros([N2, N1, 4, 4, 4])
gammaUpDownDown = zeros([N2, N1, 4, 4, 4])

dumpfile = filepath + 'gcov.h5'
datafile = h5py.File(dumpfile, "r")
for mu in xrange(4):
    for nu in xrange(4):
        gcov[:, :, mu, nu] = datafile[datafile.items()[0][0]][:][:, :, nu + 4*mu]
        
dumpfile = filepath + 'gcon.h5'
datafile = h5py.File(dumpfile, "r")
for mu in xrange(4):
    for nu in xrange(4):
        gcon[:, :, mu, nu] = datafile[datafile.items()[0][0]][:][:, :, nu + 4*mu]
        
dumpfile = filepath + 'gammaUpdowndown.h5'
datafile = h5py.File(dumpfile, "r")
for eta in xrange(4):
    for mu in xrange(4):
        for nu in xrange(4):
            gammaUpDownDown[:, :, nu, mu, eta] = \
                datafile[datafile.items()[0][0]][:][:, :, nu + 4*(mu + 4*eta)]
                
g = zeros([N2, N1])
for j in xrange(N2):
    for i in xrange(N1):
        g[j, i] = sqrt(-linalg.det(gcov[j, i, :, :]))
```

```python
# Execute this to compute the magnetic vector potential. Needed to plot the magnetic field lines.
# Warning: Computation is expensive. To compute for just the last step, set start = data_files.size - 1

from scipy.integrate import trapz

A_phi = zeros([N2, N1, data_files.size])

start = data_files.size-1
end   = data_files.size-0

for step in xrange(start, end):
    print "step = ", step
    B1 = data[:, :, 5, step]
    B2 = data[:, :, 6, step]

    for j in xrange(N2):
        for i in xrange(N1):
            A_phi[j, i, step] = trapz(g[j, :i]*B2[j, :i], dx=DX1) - trapz(g[:j, i]*B1[:j, i], dx=DX2)
```

```python
#Execute this to get hydro variables
step = -1

R_grid, THETA_grid = meshgrid(R[NG:-NG], THETA[NG:-NG])
X = R_grid*sin(THETA_grid)
Y = R_grid*cos(THETA_grid)

rho   = data[:, :, 0, step]
u     = data[:, :, 1, step]
PolGamma = 4./3
P     = (PolGamma - 1.)*u
cs    = sqrt((PolGamma-1)*P/(rho + u) )
T     = P/rho

#rhoR   = dataref[:, :, 0]
#uR     = dataref[:, :, 1]
#PR     = (PolGamma - 1.)*uR
#TR     = PR/rhoR

uGrid = data[:, :, 2:5, step]
bGrid = data[:, :, 5:8, step]
gamma = rho*0.+1
zeroVec = rho*0
alpha = sqrt(-gcon[:,:,0,0])**(-1)
for i in xrange(0, 3):
    for j in xrange(0, 3):
        gamma += gcov[:,:,i+1,j+1]*uGrid[:,:,i]*uGrid[:,:,j]
gamma = sqrt(gamma)
uCon = zeros([N2, N1, 4])
uCon[:,:,0] = gamma / alpha
for i in xrange(1, 4):
    uCon[:,:,i] = uGrid[:,:,i-1]-gamma*alpha*gcon[:,:,0,i]
uCov = zeros([N2, N1, 4])
for i in xrange(0, 4):
    for j in xrange(0, 4):
        uCov[:,:,i]+=gcov[:,:,i,j]*uCon[:,:,j]
bCon = zeros([N2, N1, 4])    
for i in xrange(0, 3):
    bCon[:,:,0]+=bGrid[:,:,i]*uCov[:,:,i+1]
for i in xrange(0, 3):
    bCon[:,:,i+1]=(bGrid[:,:,i]+bCon[:,:,0]*uCon[:,:,i+1])/uCon[:,:,0]
bCov = zeros([N2, N1, 4])
for i in xrange(0, 4):
    for j in xrange(0, 4):
        bCov[:,:,i]+=gcov[:,:,i,j]*bCon[:,:,j]
bSqr = zeros([N2, N1])
for i in xrange(0, 4):
    bSqr += bCov[:,:,i]*bCon[:,:,i]
etaVis = 1.e-2 * (cs**2)*(R_grid**(1.5))*rho
psi = data[:, :, -1, step]

divU = zeros([N2, N1])
for i in xrange(1, N1-1):
    for j in xrange(1, N2-1):
        divU[j,i]+=(g[j+1,i]*uCon[j+1,i,2]-g[j-1,i]*uCon[j-1,i,2])/2/DX2+(g[j,i+1]*uCon[j,i+1,1]-g[j,i-1]*uCon[j,i-1,1])/2/DX1
divU = divU/g
```

```python
#Execute this to get TargetPsi

TargetPsi = zeros([N2, N1])

for i in xrange(1, N1-1):
    for j in xrange(1, N2-1):
        TargetPsi[j,i] += etaVis[j,i]*divU[j,i]
        for k in xrange(0, 4):
            TargetPsi[j,i] -= 3.*etaVis[j,i]*(bCov[j,i,k]*bCon[j,i,1]*(uCon[j,i+1,k]-uCon[j,i-1,k])/2/DX1
                                        +bCov[j,i,k]*bCon[j,i,2]*(uCon[j+1,i,k]-uCon[j-1,i,k])/2/DX2)/(bSqr[j,i]+1.e-20)
            for l in xrange(0, 4):
                for m in xrange(0, 4):
                    TargetPsi[j,i] -= 3.*etaVis[j,i]*gammaUpDownDown[j,i,k,l,m]*bCov[j,i,k]*bCon[j,i,l]*uCon[j,i,m]/(bSqr[j,i]+1.e-20)
        
```
